````python
import numpy as np
from enum import IntEnum
from typing import Any, Optional, Tuple, Dict, List
import numpy.typing as npt


class Action(IntEnum):
    """
    Represents the possible actions as ints. 
    Note, that the order of the items should not be changed, 
    as they indictate the number of rotations necessary for the sliding step.
    """
    left: int = 0
    up: int = 1
    right: int = 2
    down: int = 3


class Board2048(object):
    def __init__(self, width: int = 4, height: int = 4, seed: Optional[Any] = None):
        """
        Represents a 2048 board with width and height. Actions can be performed on the board 
        with the step() function.
        
        Parameters
        ----------
        width:
            The width of the board
        height:
            The height of the board
        seed:
            The seed for the random state. Using the same seed will result in the same game to be played.
        """
        self.width = int(width)
        self.height = int(height)
        self.score = 0
            
        # Internal Variables
        self.board = None
        self.np_random = None

        self.seed(seed=seed)
        self.reset()
        

    def seed(self, seed: Optional[Any] = None) -> None:
        """
        Initialize the random state with the given seed.
        
        Parameters
        ----------
        seed:
            The seed for the random state. Using the same seed will result in the same game to be played
        """
        
        self.np_random = np.random.RandomState(seed=seed)

    def step(self, action: Action) -> Tuple[npt.NDArray[np.int64], int, bool, Dict]:        
        """
        Perform one action on the board. 
        Possible actions are [left, right, up, down] from the enum defined above. 
        Note, that impossible actions will result in no board change.
        
        This Method has to be completed!
        
        Parameters
        ----------
        action:
            The action to perform

        Returns
        -------
        board:
            The new board state
        score:
            The gained score from this move
        game_over:
            True if the game is over, False otherwise
        dict:
            Currently unused (but expected by the test code later!)
        """

        game_over = False
        score = 0
        
        
        match action:
            case Action.left:
                for y in range(board.height):
                    for x in range(board.width - 2):
                        if board.board[x][y] == board.board[x + 1][y]:
                            board.board[x][y] *= 2
                            board.board[x + 1][y] = 0
 
            case Action.up:
                pass
                            
            case Action.right:
                pass
            
            case Action.down:
                pass
        
        self.score += score
        
        return self.board, score, game_over, {}

    def game_over(self) -> bool:
        """
        Check if the game is over.
        
        This Method has to be completed!

        Returns
        -------
        game_over:
            True if the game is over, False otherwise
        """

        #TODO: INSERT YOUR CODE HERE
        

    def reset(self) -> npt.NDArray[np.int64]:
        """Clear the board and place 2 random tiles on it."""

        self.board = np.zeros((self.width, self.height), dtype=np.int64)
        self._place_random_tiles(self.board, count=2)

        return self.board

    def _sample_tiles(self, count: int = 1) -> List[int]:
        """Helper method to sample 'count' number of random tiles (value 2 or 4)."""

        choices = [2, 4]
        probs = [0.9, 0.1]

        tiles = self.np_random.choice(choices,
                                      size=count,
                                      p=probs)
        return tiles.tolist()

    def _sample_tile_locations(self, 
                               board: npt.NDArray[np.int64], 
                               count: int = 1) -> List[Tuple[int, int]]:
        """Helper method to sample 'count' random locations from the empty grid tiles."""

        zero_locs = np.argwhere(board == 0)
        zero_indices = self.np_random.choice(
            len(zero_locs), size=count)

        zero_pos = zero_locs[zero_indices]
        zero_pos = list(zip(*zero_pos))
        return zero_pos

    def _place_random_tiles(self, 
                            board: npt.NDArray[np.int64], 
                            count: int = 1) -> None:
        """
        Place 'count' random tiles on the board. 
        
        Parameters
        ----------
        board:
            The board to place the tiles on
        count:
            The number of tiles to place
        """
        if not board.all():
            tiles = self._sample_tiles(count)
            tile_locs = tuple(self._sample_tile_locations(board, count))
            board[tile_locs] = tiles

    #TODO: INSERT YOUR CODE HERE
    
````

````python
from __future__ import print_function
import ipywidgets as widgets

# change this value to test if your code is compatible for the none standard version of 2048 
board_size = 4
board = Board2048(board_size, board_size)

# RENDER CODE


color_dict = {0: '#D7D7D7', 2: '#ECE3AF', 4: '#E0D1A1', 8: '#D3B486', 16: '#F0AB79', 32: '#F18231',
              64: '#F73B0C', 128: '#DAD083', 256: '#D6C01E', 512: '#D4E80F', 1024: '#A3D829',
              2048: '#30D55A', 4096: '#B356E5', 8192: '#37E5E7', 16384: '#5117EA', 32768: '#C13695',
              65536: '#468A9C', 131072: '#9C4657'}

# Elements to Draw:
score_label = None
board_grid = None
joy_grid = None
output_debug = None

# update function. triggert by buttons. Buttons need to have action attached as dom-class
def btn_update_board(btn) -> None:
    global board
    
    with output_debug:
        action_class = btn._dom_classes[0]
        direction = Action[action_class.lower()]
        
        old_board = board.board.copy()
        # here the board model is updated according to action 'direction'
        new_board, _, game_over, _ = board.step(direction)
        
        changed = np.any(old_board != new_board)                      
        if changed:  # if the move generated any change on the grid -> update source
            update_board_grid()
            if game_over:
                print("Game Over")
                print("Re-run this cell to start again.")
            else:
                score_label.value = "Score: %d" % board.score

# create a board grid
def create_board_grid() -> widgets.GridBox:
    global board_grid, board_size
    children = [None] * board_size**2
    for i in range(board_size):
        for j in range(board_size):
            id = i * board_size + j
            btn = widgets.Button(description="0", layout=widgets.Layout(width='auto', height='auto', border='10px solid #B8AD9A', margin="-1px"),
                         style=widgets.ButtonStyle(button_color='#D7D7D7', text_color='#D7D7D7'))
            button_clickable = False
            if j == 0 and i > 0 and i < board_size - 1:
                btn.add_class("LEFT") # not very good solution, I need a visible name and some kind of class, but not in DOM
                button_clickable = True
            elif j == board_size - 1 and i > 0 and i < board_size - 1:
                btn.add_class("RIGHT")
                button_clickable = True
            elif i == 0 and j > 0 and j < board_size - 1:
                btn.add_class("UP")
                button_clickable = True
            elif i == board_size - 1 and i > 0:
                btn.add_class("DOWN")
                button_clickable = True
                
            if button_clickable:
                btn.on_click(btn_update_board)
            
            children[id] = btn
    return widgets.GridBox(children=children,
        layout=widgets.Layout(
            grid_template_columns='100px ' * board_size,
            grid_template_rows='100px ' * board_size,
            grid_gap='0px 0px'),
        style=widgets.Style(background_color='black')
       )

# update board rendering: synchronize model with view
def update_board_grid() -> None:
    global board, board_grid, board_size
    for i in range(board_size):
        for j in range(board_size):
            number = board.board[i,j]
            bg_color = color_dict[number]
            text_color = 'black' if number != 0 else '#D7D7D7'
            board_grid.children[i * board_size + j].style.button_color = bg_color
            board_grid.children[i * board_size + j].style.text_color = text_color
            board_grid.children[i * board_size + j].description = "%d" % number

# create a 
def create_joy_button(name, viz_name = None) -> widgets.Button:
    if viz_name is None:
        viz_name = name
    btn = widgets.Button(description=viz_name, layout=widgets.Layout(width='auto', height='auto'))
    btn.on_click(btn_update_board)
    btn.add_class(name) # not very good solution, I need a visible name and some kind of class, but not in DOM
    return btn

# INIT
score_label = widgets.Label(value="Score: %d" % board.score)
board_grid = create_board_grid()
update_board_grid()
output_debug = widgets.Output()
joy_grid = widgets.GridspecLayout(2,3, width="300px", height="100px")
joy_grid[0, 1] = create_joy_button("UP")
joy_grid[1, 0] = create_joy_button("LEFT")
joy_grid[1, 2] = create_joy_button("RIGHT")
joy_grid[1, 1] = create_joy_button("DOWN")

# RENDER
display(score_label)
display(board_grid)
display(joy_grid)
display(output_debug)
````
